#### Ниже пробую создавать 2 выборки , чтобы поиграться с  их значениями. И поэксперементировать с стат.тестами.

```python
np.random.seed(42)
n1, n2 = 10000, 10000
mu1, mu2 = 34, 33
sigma1, sigma2 = 10, 10
```

```python
sample_1 = np.random.normal(loc = mu1, scale = sigma1, size = n1)
sample_2 = np.random.normal(loc = mu2, scale = sigma2, size = n2)

print(f'Sample_1: среднее = {np.mean(sample_1)}, стандартное отклонение = {np.std(sample_1)}, медиана = {np.median(sample_1)}')
print(f'Sample_2: среднее = {np.mean(sample_2)}, стандартное отклонение = {np.std(sample_2)}, медиана = {np.median(sample_2)}')
```

```python
sns.histplot(sample_1, color = 'blue', label = 'sample_1')
sns.histplot(sample_2, color = 'yellow', label = 'sample_2')
plt.axvline(34, color = 'red', linestyle = '--')
plt.axvline(33, color = 'brown' , linestyle =':')
```

```python
sns.boxplot(data = [sample_1, sample_2])
plt.xticks([0, 1], ['Sample_1', 'Sample_2'])
plt.title('Ящик с усами для наших сгенерированных выборок\n')
```

```python
ttest_ind(sample_1, sample_2)
```

```python
mannwhitneyu(sample_1, sample_2)
```

```python
a = bootstrap((sample_1,), np.median)
a
```

```python
sns.histplot(a.bootstrap_distribution, bins = 30)
```

```python
b = bootstrap((sample_2,), np.median)
b
```

```python
sns.histplot(b.bootstrap_distribution, bins = 30)
```
